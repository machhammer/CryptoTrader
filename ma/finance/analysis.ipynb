```python
import datetime as dt
from dateutil.relativedelta import relativedelta
import data_provider.DataReader as data_provider
from trading.SMAStrategy import SmaCross
from trading.MyStrategy import MyStrategy
import config as cfg
import yahoo_fin.stock_info as yf
import pandas as pd
import backtrader as bt
```

```python
end_date = dt.datetime.now()
start_date = end_date - relativedelta(months=6)
data = data_provider.instatiate("yahoo").historic_price_data("xrp-usd", start = start_date, end = end_date)
#data = yf.get_data("xrp-usd", start_date = start_date, end_date = end_date, index_as_date = True, interval = '1d')
data
```

```python
cerebro = bt.Cerebro()
cerebro.addstrategy(SmaCross)
cerebro.adddata(bt.feeds.PandasData(dataname=data))
start_portfolio_value = cerebro.broker.getvalue()
cerebro.run()
```

```python
end_portfolio_value = cerebro.broker.getvalue()
pnl = end_portfolio_value - start_portfolio_value
print(f'Starting Portfolio Value: {start_portfolio_value:2f}')
print(f'Final Portfolio Value: {end_portfolio_value:2f}')
print(f'PnL: {pnl:.2f}')
```
